```python
# prompt: Faça uma lista de nome "Quantitativas" utilizando as seguintes colunas do dataset: Release Year, Monthly Listeners , Total Streams, Total Hours Streamed , Avg Stream Duration, Streams Last 30 Days,  Skip Rate. Imprima a lista com o nome dela no começo.

Quantitativas = ['Release Year', 'Monthly Listeners', 'Total Streams', 'Total Hours Streamed', 'Avg Stream Duration', 'Streams Last 30 Days', 'Skip Rate']
print("Quantitativas:", Quantitativas)
```

```python
# prompt: Faça uma lista de nome "Qualitativas" adicionando as seguintes colunas do dataset: Country, Artist, Album, Genre, Platform Type. Imprima a lista com o nome dela no começo.

Qualitativas = ['Country', 'Artist', 'Album', 'Genre', 'Platform Type']
print("Qualitativas:", Qualitativas)
```

```python
# prompt: Crie uma tabela com as duas colunas: "Qualitativas" e "Quantitativas". Utilize as variáveis das listas criadas anteriormente para preenchê-la, se as listas não tiverem o mesmo tamanho, adicione variáveis vazias para igualar. Exiba a tabela criada

import pandas as pd
# Create a list of lists to represent the table data
table_data = []

# Determine the maximum length of the two lists
max_len = max(len(Qualitativas), len(Quantitativas))

# Iterate through the lists and create rows for the table
for i in range(max_len):
  row = []
  if i < len(Qualitativas):
    row.append(Qualitativas[i])
  else:
    row.append("")  # Add an empty string if Qualitativas is shorter

  if i < len(Quantitativas):
    row.append(Quantitativas[i])
  else:
    row.append("")  # Add an empty string if Quantitativas is shorter

  table_data.append(row)

# Create the Pandas DataFrame
df_table = pd.DataFrame(table_data, columns=["Qualitativas", "Quantitativas"])

# Display the table
df_table
```
